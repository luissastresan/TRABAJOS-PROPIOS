3. Write a program that asks the user for her/his name and the current date. Then, store the information in a single line in a new text file named 'second_text.txt'. The line in the text file should read like: 'I am ```Name```, and today is ```Date```'. Close the file.

````python
name = input("Introduce tu nombre: ")
date = input("Introduce la fecha de hoy: ")

texto = f"I am {name}, and today is {date}"

fichero2 = open("second_text.txt", "w")
fichero2.write(texto)
fichero2.close()
````

4. Open the file created in exercise 3 and read its contents. Print them on the screen.

````python
with open("second_text.txt", "r") as fichero:
    for linea in fichero:
        print(linea)
````

````python
fichero.close()
del(fichero)
````
